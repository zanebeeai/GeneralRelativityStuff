```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
from matplotlib import patheffects

# Setting the font properties globally
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.size'] = 12
mpl.rcParams['text.color'] = 'white'

# Example data for X, Y, and Gamma_22
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Gamma_22 = np.exp(-0.1*(X**2 + Y**2))

fig = plt.figure(figsize=(180, 6))
# fig.patch.set_facecolor('black')

ax = fig.add_subplot(111, projection='3d')
# ax.patch.set_facecolor('black')

# Plotting the surface
surf = ax.plot_surface(X, Y, -Gamma_22, cmap='plasma', edgecolor='none', linewidth=0, alpha=0.6, rstride=1, cstride=1)
fig.colorbar(surf)

# Setting labels with shadow for better visibility
def add_shadow(text, shadow_color='black'):
    text_path_effects = [mpl.patheffects.withStroke(linewidth=3, foreground=shadow_color)]
    text.set_path_effects(text_path_effects)

label_x = ax.set_xlabel('X', color="white")
label_y = ax.set_ylabel('Y', color="white")
label_z = ax.set_zlabel(r'$\gamma_{22}$', color="white")

add_shadow(label_x)
add_shadow(label_y)
add_shadow(label_z)

# Adjust view angle
ax.view_init(elev=40, azim=45)

# Adjust grid line visibility and color
ax.xaxis._axinfo["grid"].update(color = "gray", linestyle = "--", linewidth=0.5)
ax.yaxis._axinfo["grid"].update(color = "gray", linestyle = "--", linewidth=0.5)
ax.zaxis._axinfo["grid"].update(color = "gray", linestyle = "--", linewidth=0.5)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
from matplotlib import patheffects

# Setting the font properties globally
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.size'] = 14
mpl.rcParams['text.color'] = 'white'  # Ensure this is set correctly

# Example data for X, Y, and Gamma_22
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Gamma_22 = np.exp(-0.1 * (X**2 + Y**2))  # Placeholder function

fig = plt.figure()
fig.patch.set_facecolor('black')

ax = fig.add_subplot(111, projection='3d')
ax.patch.set_facecolor('black')

# Plotting the surface
surf = ax.plot_surface(X, Y, -Gamma_22, cmap='plasma', edgecolor='none', linewidth=0, alpha=0.5)
fig.colorbar(surf)

# Setting labels with enhanced visibility
ax.set_xlabel('X', color="white", fontsize=14)
ax.set_ylabel('Y', color="white", fontsize=14)
ax.set_zlabel(r'$\gamma_{22}$', color="white", fontsize=14)

# Add shadows to make text stand out
for label in [ax.xaxis.label, ax.yaxis.label, ax.zaxis.label]:
    label.set_path_effects([patheffects.withStroke(linewidth=3, foreground='black')])

# Adjust view angle
ax.view_init(elev=40, azim=45)

# Enhance grid line visibility
ax.xaxis._axinfo["grid"].update(color="white", linestyle="--", linewidth=1)
ax.yaxis._axinfo["grid"].update(color="white", linestyle="--", linewidth=1)
ax.zaxis._axinfo["grid"].update(color="white", linestyle="--", linewidth=1)

plt.show()
```
