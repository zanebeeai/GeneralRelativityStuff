```python
import sympy as sp
def get_g00(lineElement, subs=None):
    
    lineElement=sp.expand(lineElement)
    lineElement = lineElement.subs(subs)
    # Define the coordinate symbols
    lam = sp.symbols('lambda')
    t,x,y,z = sp.symbols('t x y z')
    dt = sp.symbols('dt')
    
    # Calculate the coefficient for dt**2
    g00_coeff = lineElement.coeff(dt**2)
    
    # Substitute any additional simplifications
    if subs:
        g00_coeff = g00_coeff.subs(subs).simplify()
    
    return g00_coeff
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm  # Import colormap

def plotGrav(g00, addArgs):
    xGspan, yGspan = np.linspace(-5, 5, 200).astype(float),np.linspace(-5, 5, 200).astype(float)
    X, Y = np.meshgrid(xGspan, yGspan)
    
    symbols = [i for i in sp.symbols('t x y z')]  # Default symbols for coordinates
    values = [0, X, Y, 0.5]              # Default values for x, y, and a placeholder for z

    # Extend symbols and values from addArgs
    for symbol, value in addArgs:
        symbols.append(symbol)
        values.append(value)

    gravPot = sp.lambdify(symbols, -0.5*(g00+1), "numpy")
    pots=gravPot(*values)

    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    surface = ax.plot_surface(X, Y, -pots, cmap='viridis', edgecolor='none')
    fig.colorbar(surface)

    ax.set_title('3D Mesh of Gravitational Potential')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('Potential')
    ax.view_init(elev=30, azim=-30)

    plt.show()
```

```python
rs, R, lam = sp.symbols("r_s, R, lambda")
G, M, c = sp.symbols("G M c")

t,x,y,z = sp.symbols('t x y z')

dt,dx,dy,dz = sp.symbols('dt dx dy dz')

lineElement = (1-rs/4/R)**2/(1+rs/4/R)**2 *dt**2 -(1+rs/4/R)**4 *(dx**2+dy**2+dz**2)
display(lineElement)

# order of substitutions matter!!!
subs= [
    (R, sp.sqrt(x**2+y**2+z**2)),
    (rs, 2*G*M/(c**2))
]
addArgs = [
    (G, 1),
    (M, 1),
    (c, 1)
]
display(get_g00(lineElement, subs))
plotGrav(get_g00(lineElement, subs), addArgs)
```

```python
f_r, vs, r, sigma, xs, R = sp.symbols("f_r vs r sigma xs R")

t,x,y,z = sp.symbols('t x y z')

dt,dx,dy,dz = sp.symbols('dt dx dy dz')

lineElement = -dt**2 + (dx-vs*f_r*dt)**2 + dy**2 + dz**2
display(lineElement)

# order of substitutions matter!!!
subs= [
    (f_r, (sp.tanh(sigma * (r + R)) - sp.tanh(sigma * (r - R))) / (2 * sp.tanh(sigma * R))),
    (r, sp.sqrt((x - xs)**2 + y**2 + z**2)),
    (xs, vs*t)  # since its steady state
]

# Additional arguments
addArgs = [
    (vs, -10),
    (sigma, 1.0),
    (R, 4.0)
]
display(get_g00(lineElement, subs))
plotGrav(get_g00(lineElement, subs), addArgs)
```
